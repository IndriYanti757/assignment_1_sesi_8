### Advanced Visualization

```python
import numpy as np
import pandas as pd
from PIL import Image
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

# cek version matplotlib
print('Matplotlib version: ', mpl.__version__)
```

```python
total_crime_per_month = df.groupby(['year', 'month'])['value'].sum().reset_index()
```

```python
# Membuat plot line menggunakan data total kejahatan per bulan
plt.figure(figsize=(10,6))  # Mengatur ukuran plot
plt.plot(total_crime_per_month['month'], total_crime_per_month['value'], marker='o', linestyle='-')  # Menambahkan data untuk plot
plt.title('Total Crime Trends')  # Judul plot
plt.xlabel('Month')  # Label sumbu x
plt.ylabel('Total Crime Value')  # Label sumbu y
plt.grid(True)  # Menambahkan grid
plt.xticks(total_crime_per_month['month'])  # Menyertakan label bulan pada sumbu x
plt.show()  # Menampilkan plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
# Menghitung total nilai kejahatan untuk setiap bulan
total_crime_per_month = df.groupby(['year', 'month'])['value'].sum().reset_index()

# Menyertakan data yang dibutuhkan untuk Waffle Chart
data = {str(month): value for month, value in zip(total_crime_per_month['month'], total_crime_per_month['value'])}

# Membuat Waffle Chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=12,
    values=data,
    title={'label': 'Total Crime Trends', 'loc': 'left'},
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'fontsize': 10},
    icons='child',
    icon_style='solid',
    icon_legend=True,
    figsize=(10, 6)
)

plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Batasi jumlah data yang digunakan (misalnya, 1000 baris)
limited_df = df.head(1000)
```

```python
# Menggabungkan teks dari kolom borough, major_category, dan minor_category untuk data terbatas
text = ' '.join(limited_df['borough']) + ' ' + ' '.join(limited_df['major_category']) + ' ' + ' '.join(limited_df['minor_category'])
```

```python
# Membuat objek WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Menampilkan Word Cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Crime Data')
plt.axis('off')  # Menghilangkan sumbu x dan y
plt.show()
```

```python
from geopy.geocoders import Nominatim
```

```python
# Baca sebagian data dari file CSV ke dalam DataFrame
df = pd.read_csv("london_crime_by_lsoa.csv", nrows=100)
```

```python
# Membuat peta dasar menggunakan Folium
world_map = folium.Map(location=[51.5074, -0.1278], zoom_start=10) 
```

```python
# Membuat objek geocoder
geolocator = Nominatim(user_agent="london_explorer")

# Menambahkan marker untuk setiap borough pada peta
for index, row in df.iterrows():
    location = geolocator.geocode(row['borough'])
    if location:
        folium.Marker(location=[location.latitude, location.longitude], popup=row['borough']).add_to(world_map)

# Menampilkan peta
world_map
```
